Write a Python function named `is_palindrome_number` that takes an integer as input and returns `True` if the number is a palindrome, and `False` otherwise. A palindrome number reads the same forwards and backwards.

```python
def is_palindrome_number(number):
    """
    Checks if a given integer is a palindrome number.

    A palindrome number reads the same forwards and backwards.

    Args:
        number (int): The integer to check.

    Returns:
        bool: True if the number is a palindrome, False otherwise.
    """
    # Convert the number to a string to easily compare characters
    num_str = str(number)

    # Compare the string with its reverse
    return num_str == num_str[::-1]
```

### Explanation of the `is_palindrome_number` function:

1.  **Convert to String**: The input `number` (an integer) is first converted into a string using `str(number)`. This is because it's easier to manipulate and reverse strings than integers.
2.  **Reverse String**: `num_str[::-1]` creates a reversed version of the `num_str`. The `[::-1]` slicing syntax is a common Python idiom for reversing sequences.
3.  **Compare**: The original string `num_str` is then compared with its reversed version. If they are identical, the number is a palindrome, and the function returns `True`; otherwise, it returns `False`.

Write a Python function named `calculate_factorial` that takes an integer `n` as input and returns its factorial.

**Example:**
Input: `5`  Output: `120`

```python
def calculate_factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    elif n == 0:
        return 1
    else:
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i
        return factorial
```

### Explanation of the `calculate_factorial` function:

1.  **Input Validation**: The function first checks if the input `n` is a negative number. If it is, a `ValueError` is raised because factorials are not defined for negative integers.
2.  **Base Case (n=0)**: If `n` is 0, the function returns 1, as the factorial of 0 is 1.
3.  **Iterative Calculation**: For positive integers, the function initializes a `factorial` variable to 1. It then iterates from 1 up to `n` (inclusive), multiplying `factorial` by each number in the loop. Finally, the calculated `factorial` is returned.

```python
# Test cases for the calculate_factorial function
print(f"Factorial of 5: {calculate_factorial(5)} (Expected: 120)")
print(f"Factorial of 0: {calculate_factorial(0)} (Expected: 1)")
print(f"Factorial of 1: {calculate_factorial(1)} (Expected: 1)")
print(f"Factorial of 3: {calculate_factorial(3)} (Expected: 6)")

# Test with a negative number (should raise ValueError)
try:
    calculate_factorial(-1)
except ValueError as e:
    print(f"Error for negative input: {e}")
```

Few-Shot Prompting (Armstrong Number Function):
Prompt: "Write a Python function named `is_armstrong_number` that takes an integer `num` as input and returns `True` if the number is an Armstrong number, and `False` otherwise. An Armstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits.

Examples:
 Input: 153  Output: Armstrong Number
 Input: 370  Output: Armstrong Number
 Input: 123  Output: Not an Armstrong Number"

```python
def is_armstrong_number(num):
    if not isinstance(num, int):
        raise TypeError("Input must be an integer.")
    if num < 0:
        # Armstrong numbers are typically defined for non-negative integers
        return False

    num_str = str(num)
    n_digits = len(num_str)
    sum_of_powers = 0

    for digit_char in num_str:
        digit = int(digit_char)
        sum_of_powers += digit ** n_digits

    return sum_of_powers == num
```

3.  **Iterate through digits:** Each digit needs to be extracted, converted back to an integer, and then raised to the power of `n_digits`.
4.  **Sum the results:** The individual powered digits are summed up.
5.  **Compare the sum to the original number:** This is the final check.

The examples enhance accuracy by establishing a clear pattern. Without examples, an AI might misinterpret the definition (e.g., raise to the power of 3 always, or sum digits without powers). The examples make the 'number of digits' part of the definition explicit in practice.

4.  **Count Digits**: The number of digits (`n_digits`) is obtained using `len(num_str)`.
5.  **Calculate Sum of Powers**: A loop iterates through each character in `num_str`. Each character is converted to an integer (`digit`), and then `digit` is raised to the power of `n_digits` and added to `sum_of_powers`.

You are a Python programming expert. Your task is to design and implement a highly optimized Python function for classifying a given integer as 'prime', 'composite', or 'neither'.

**Function Name:** `classify_number`

**Input:** An integer `num`.

**Output:** A string: 'prime', 'composite', or 'neither'.

**Instructions and Constraints:**

1.  **Definition of Classification:**
    *   'prime': A natural number greater than 1 that has no positive divisors other than 1 and itself.
    *   'composite': A natural number greater than 1 that is not prime.
    *   'neither': Numbers less than or equal to 1 (e.g., 0, 1, negative integers, non-integers).

2.  **Input Validation:**
    *   The function must robustly handle non-integer inputs by raising a `TypeError`.
    *   For integers `num <= 1`, the function should return 'neither'.

3.  **Optimization for Efficiency:**
    *   The logic for checking primality should be as efficient as possible, especially considering potentially large input numbers. Think about mathematical properties and common optimization techniques for primality tests (e.g., checking divisors only up to the square root of the number, handling even numbers separately).

4.  **Example:**
    *   Input: 17  Output: prime
    *   Input: 20  Output: composite
    *   Input: 1  Output: neither

```python
import math

def classify_number(num):
    # Input Validation: Handle non-integer inputs
    if not isinstance(num, int):
        raise TypeError("Input must be an integer.")

    # Classification for num <= 1
    if num <= 1:
        return 'neither'

    # Optimization: Handle even numbers separately for primality test
    if num == 2:
        return 'prime'
    if num % 2 == 0:
        return 'composite'

    # Optimization: Check divisors only up to the square root of num
    # We start from 3 and increment by 2 (only odd divisors needed)
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return 'composite'

    # If no divisors found, it's a prime number
    return 'prime'
```

### Explanation of the `classify_number` function:

1.  **Input Validation (`isinstance(num, int)`):**
    *   The function first checks if the input `num` is an integer. If not, it raises a `TypeError` as specified, ensuring robust handling of invalid input types.

2.  **Handle 'neither' cases (`num <= 1`):**
    *   Numbers less than or equal to 1 (including negative integers, 0, and 1) are classified as 'neither' based on the definition of prime and composite numbers.

3.  **Optimization for Efficiency:**
    *   **Even Numbers:** The function explicitly handles the number `2` as 'prime'. For any other even number (`num % 2 == 0`), it's immediately classified as 'composite'. This step is a significant optimization as it reduces the number of checks needed for subsequent loops by eliminating all even potential divisors.
    *   **Square Root Limit:** For odd numbers greater than 2, the primality test checks for divisors only up to the square root of `num`. If a number `n` has a divisor `d` greater than `sqrt(n)`, then it must also have a divisor `n/d` which is less than `sqrt(n)`. Therefore, checking up to the square root is sufficient.
    *   **Odd Divisors Only:** The loop for checking divisors starts from `3` and increments by `2` (`range(3, ..., 2)`). This ensures that only odd numbers are checked as potential divisors, further improving efficiency since even divisors (other than 2) have already been excluded.

4.  **Classification Logic:**
    *   If the number passes all the primality checks (i.e., it's not `neither`, not `composite` due to being even, and has no odd divisors up to its square root), it is classified as 'prime'.

```python
# Test cases for the classify_number function

# Examples from the prompt
print(f"17: {classify_number(17)} (Expected: prime)")
print(f"20: {classify_number(20)} (Expected: composite)")
print(f"1: {classify_number(1)} (Expected: neither)\n")

# Other 'neither' cases
print(f"0: {classify_number(0)} (Expected: neither)")
print(f"-5: {classify_number(-5)} (Expected: neither)\n")

# Other prime numbers
print(f"2: {classify_number(2)} (Expected: prime)")
print(f"3: {classify_number(3)} (Expected: prime)")
print(f"23: {classify_number(23)} (Expected: prime)")
print(f"97: {classify_number(97)} (Expected: prime)\n")

# Other composite numbers
print(f"4: {classify_number(4)} (Expected: composite)")
print(f"9: {classify_number(9)} (Expected: composite)")
print(f"25: {classify_number(25)} (Expected: composite)")
print(f"100: {classify_number(100)} (Expected: composite)\n")

# Test with a large prime number
print(f"1000003: {classify_number(1000003)} (Expected: prime)\n") # A large prime

# Test with non-integer inputs (should raise TypeError)
try:
    classify_number(10.5)
except TypeError as e:
    print(f"Error for float input: {e}")

try:
    classify_number("hello")
except TypeError as e:
    print(f"Error for string input: {e}")

try:
    classify_number([1, 2])
except TypeError as e:
    print(f"Error for list input: {e}")
```

```python
import math

def is_perfect_number(num):
    # First, check if the input is an integer.
    if not isinstance(num, int):
        return False

    # Then, handle cases where num is not a positive integer
    if num <= 0:
        return False # Perfect numbers are positive integers

    # Explicitly handle 1: it has no proper positive divisors, so sum is 0. Thus, not perfect.
    if num == 1:
        return False

    # Find proper divisors and sum them
    sum_divisors = 1 # Start with 1 as it's always a divisor for num > 1
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            sum_divisors += i
            if i * i != num: # Avoid adding the square root twice for perfect squares
                sum_divisors += num // i

    return sum_divisors == num
```

### Explanation of the `is_perfect_number` function:

1.  **Input Validation**: The function first checks if `num` is less than or equal to 0 or not an integer. Perfect numbers are defined as positive integers, so these inputs immediately return `False`.
2.  **Initialize Sum of Divisors**: `sum_divisors` is initialized to 1 because 1 is a proper divisor of every positive integer.
3.  **Calculate Divisors Efficiently**: The loop iterates from 2 up to the square root of `num`. This is an optimization technique because if `i` is a divisor of `num`, then `num // i` is also a divisor.
    *   If `num % i == 0`, then `i` is a divisor, and it's added to `sum_divisors`.
    *   `num // i` is also a divisor. To avoid double-counting in the case of perfect squares (where `i * i == num`), a condition `if i * i != num` is used before adding `num // i`.
4.  **Compare Sum with Number**: Finally, the function returns `True` if `sum_divisors` is equal to `num`, indicating it's a perfect number; otherwise, it returns `False`.
