# **Task Description-1**


• Zero-shot: Prompt AI with only the instruction. Write a Python function to determine
whether a given number is prime

Expected Output-1

• A basic Python function to check if a number is prime, demonstrating correct logical
conditions without relying on examples or additional contex



> Zero-Shot Prompt:
Write a Python function named is_prime_number that takes an integer as input and returns True if the number is prime, and False otherwise. Ensure the function handles edge cases like numbers less than 2



```python
def is_prime_number(num):
    """
    Checks if a given integer is a prime number.

    A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

    Args:
        num (int): The integer to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Test cases
print(f"Is 2 prime? {is_prime_number(2)}")
print(f"Is 1 prime? {is_prime_number(1)}")
print(f"Is 0 prime? {is_prime_number(0)}")
print(f"Is 7 prime? {is_prime_number(7)}")
print(f"Is 9 prime? {is_prime_number(9)}")
print(f"Is 11 prime? {is_prime_number(11)}")
print(f"Is 10 prime? {is_prime_number(10)}")
```

This code checks if a number is prime. A prime number is only divisible by 1 and itself.

* If the number is less than 2, it’s not prime.

* The loop checks if any number from 2 to the square root of the number divides it without a remainder.

* If it finds one, the number isn’t prime. If not, it is.

Test cases check different numbers to see if they're prime.

# **Task Description-2**


• One-shot: Provide one example: Input: [1, 2, 3, 4], Output: 10 to help AI generate a
function that calculates the sum of elements in a list.

Expected Output-2

• A correct conversion function guided by the single example.



> One-Shot Prompt: Write a Python function called sum_list_elements that takes a list of numbers as input and returns their sum. Your function should be guided by the following example:

Input: [1, 2, 3, 4] Output: 10



```python
def sum_list_elements(numbers):
    """
    Calculates the sum of all elements in a list of numbers.

    The function is guided by the example: Input: [1, 2, 3, 4], Output: 10.

    Args:
        numbers (list): A list of numerical elements.

    Returns:
        int or float: The sum of the elements in the list.
    """
    total_sum = 0
    for number in numbers:
        total_sum += number
    return total_sum

# Test case based on the provided example
example_input = [1, 2, 3, 4]
example_output = sum_list_elements(example_input)
print(f"Input: {example_input}, Output: {example_output}")

# Additional test cases
print(f"Input: [5, 5, 5], Output: {sum_list_elements([5, 5, 5])}")
print(f"Input: [], Output: {sum_list_elements([])}")
print(f"Input: [10], Output: {sum_list_elements([10])}")
```

This code adds up all the numbers in a list and gives the total.

* It starts with a total of 0.

* Then, it goes through each number in the list and adds it to the total.

* After going through all numbers, it returns the total sum.

Test cases check different lists to see the sum.

# **Task Description-3**


• Few-shot: Give 2–3 examples to create a function that extracts digits from an
alphanumeric string.

Expected Output-3

• Accurate function that returns only the digits from alphanumeric string.



> Few-Shot Prompt: Write a Python function called extract_digits that takes an alphanumeric string as input and returns a string containing only the digits from the input string. The function should be robust and handle various input formats.

Few examples:

Input: extract_digits("abc123def456") Output: "123456"

Input: extract_digits("no_digits_here") Output: ""

Input: extract_digits("Value: $100.50") Output: "10050"



```python
import re

def extract_digits(alphanum_string):
    """
    Extracts only the digits from an alphanumeric string.

    The function is guided by the examples:
    - Input: "abc123def456", Output: "123456"
    - Input: "no_digits_here", Output: ""
    - Input: "Value: $100.50", Output: "10050"

    Args:
        alphanum_string (str): The input alphanumeric string.

    Returns:
        str: A string containing only the digits from the input string.
    """
    return re.sub(r'[^0-9]', '', alphanum_string)

# Test cases based on the provided examples
print(f"Input: 'abc123def456', Output: '{extract_digits("abc123def456")}'")
print(f"Input: 'no_digits_here', Output: '{extract_digits("no_digits_here")}'")
print(f"Input: 'Value: $100.50', Output: '{extract_digits("Value: $100.50")}'")

# Additional test cases
print(f"Input: '123 Test 45', Output: '{extract_digits("123 Test 45")}'")
print(f"Input: 'Mixed12Chars345', Output: '{extract_digits("Mixed12Chars345")}'")
```

This code extracts only the digits from a string, removing everything else.

* The function uses a regular expression (re.sub) to remove anything that's not a number (0-9).

* It returns a string with just the digits from the input.

Test cases show how it works with different types of input strings.

# **Task Description-4**


• Compare zero-shot vs few-shot prompting for generating a function that counts the
number of vowels in a string.

Expected Output-4

• Output comparison + student explanation on how examples helped the model.



> Zero-Shot Prompt: Write a Python function called count_vowels that takes a string as input and returns the total number of vowels (a, e, i, o, u, case-insensitive) present in the string.



```python
def count_vowels(input_string):
    """
    Counts the total number of vowels (a, e, i, o, u, case-insensitive)
    present in the string.

    Args:
        input_string (str): The string to analyze.

    Returns:
        int: The total number of vowels.
    """
    vowels = "aeiou"
    vowel_count = 0
    for char in input_string:
        if char.lower() in vowels:
            vowel_count += 1
    return vowel_count

# Test cases for the zero-shot prompt
print(f"'Hello World' has {count_vowels('Hello World')} vowels.")
print(f"'Python Programming' has {count_vowels('Python Programming')} vowels.")
print(f"'One Two Three' has {count_vowels('One Two Three')} vowels.")
print(f"'Rhythm' has {count_vowels('Rhythm')} vowels.")
print(f"'AEIOU' has {count_vowels('AEIOU')} vowels.")
```

> Few-Shot Prompt: Write a Python function called count_vowels_with_examples that takes a string as input and returns the total number of vowels (a, e, i, o, u, case-insensitive) present in the string. The function should be guided by the following examples:

Input: count_vowels_with_examples("Hello World") Output: 3

Input: count_vowels_with_examples("Python") Output: 2

Input: count_vowels_with_examples("AI is amazing!") Output: 6



```python
def count_vowels_with_examples(input_string):
    """
    Counts the total number of vowels (a, e, i, o, u, case-insensitive)
    present in the string, guided by examples.

    Args:
        input_string (str): The string to analyze.

    Returns:
        int: The total number of vowels.
    """
    vowels = "aeiou"
    # Using a generator expression with sum() for a more concise and efficient approach
    return sum(1 for char in input_string if char.lower() in vowels)

# Test cases from the prompt
print(f"Input: 'Hello World', Output: {count_vowels_with_examples('Hello World')}") # Expected: 3
print(f"Input: 'Python', Output: {count_vowels_with_examples('Python')}")             # Expected: 2
print(f"Input: 'AI is amazing!', Output: {count_vowels_with_examples('AI is amazing!')}") # Expected: 6

# Additional test cases for robustness
print(f"Input: 'Eiffel Tower', Output: {count_vowels_with_examples('Eiffel Tower')}") # Expected: 5
print(f"Input: 'Rhythm', Output: {count_vowels_with_examples('Rhythm')}")             # Expected: 0
print(f"Input: '', Output: {count_vowels_with_examples('')}")                    # Expected: 0
```

**Zero-Shot vs Few-Shot Prompting:**

*Zero-Shot Prompt:*

Prompt: "Write a Python function to count vowels in a string."

Explanation: The model generates the solution based on its general knowledge without examples. It works fine for basic tasks but might miss edge cases or optimization.

Output: A working function that counts vowels but could be less efficient.

----

*Few-Shot Prompt:*

Prompt: "Write a Python function to count vowels, guided by these examples."

Example inputs like "Hello World" → 3, "Python" → 2.

Explanation: The model uses the provided examples to better understand how to handle edge cases, optimize the solution, and improve accuracy.

Output: A more efficient, concise function that handles cases like punctuation and mixed case letters.

---

**Student Explanation:**

Zero-Shot: When the prompt doesn’t give examples, the model just relies on its general knowledge. It can still create a working function, but it might not be optimized or handle all edge cases, like spaces or punctuation.

Few-Shot: The examples help the model understand exactly how it should behave in different situations. For example, it learns how to handle spaces and punctuation properly. This guidance makes the function more efficient and accurate, covering more cases.

# **Task Description-5**


• Use few-shot prompting with 3 sample inputs to generate a function that determines
the minimum of three numbers without using the built-in min() function.

Expected Output-5

• A function that handles all cases with correct logic based on example patterns.


>Few-Shot Prompt: Write a Python function called find_minimum_of_three that takes three numbers (num1, num2, num3) as input and determines the smallest among them. This function should not use Python's built-in min() function. Instead, it should use conditional logic to compare the numbers.

Examples:

Input: find_minimum_of_three(5, 2, 8) Output: 2

Input: find_minimum_of_three(10, 10, 7) Output: 7

Input: find_minimum_of_three(4, 9, 4) Output: 4



```python
def find_minimum_of_three(num1, num2, num3):
    """
    Determines the minimum of three numbers without using the built-in min() function.

    Args:
        num1 (int or float): The first number.
        num2 (int or float): The second number.
        num3 (int or float): The third number.

    Returns:
        int or float: The smallest of the three numbers.
    """
    minimum = num1
    if num2 < minimum:
        minimum = num2
    if num3 < minimum:
        minimum = num3
    return minimum

# Test cases from the prompt
print(f"Input: find_minimum_of_three(5, 2, 8), Output: {find_minimum_of_three(5, 2, 8)}") # Expected: 2
print(f"Input: find_minimum_of_three(10, 10, 7), Output: {find_minimum_of_three(10, 10, 7)}") # Expected: 7
print(f"Input: find_minimum_of_three(4, 9, 4), Output: {find_minimum_of_three(4, 9, 4)}") # Expected: 4

# Additional test cases for robustness
print(f"Input: find_minimum_of_three(15, 20, 12), Output: {find_minimum_of_three(15, 20, 12)}") # Expected: 12
print(f"Input: find_minimum_of_three(-5, -10, -2), Output: {find_minimum_of_three(-5, -10, -2)}") # Expected: -10
print(f"Input: find_minimum_of_three(3.14, 2.71, 1.618), Output: {find_minimum_of_three(3.14, 2.71, 1.618)}") # Expected: 1.618
```

This code finds the smallest number among three numbers without using the min() function.

* It starts by assuming the first number is the smallest.

* Then, it compares the second and third numbers to see if either is smaller.

* The smallest number is returned as the result.

Test cases check different sets of numbers to find the smallest one.
