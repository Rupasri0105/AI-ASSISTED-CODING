**Task 1: Cleaning Sensor Data**

❖ Scenario:

❖ You are cleaning IoT sensor data where negative values are invalid.

❖ Task:

Use Gemini in Colab to generate a function that filters out all negative

````python
def get_negative_numbers(data_list):
    """Extracts all negative numbers from a given list."""
    return [num for num in data_list if num < 0]

# Get user input for sensor data
user_input = input("Enter a comma-separated list of numbers (e.g., 10, -5, 20, 0, -15): ")

# Convert the input string to a list of numbers
try:
    sensor_data_original = [int(x.strip()) for x in user_input.split(',')]
except ValueError:
    print("Invalid input. Please enter integer numbers separated by commas.")
    sensor_data_original = []

print(f"Original List: {sensor_data_original}")

# Apply the function to get negative numbers
negative_numbers_list = get_negative_numbers(sensor_data_original)

print(f"Extracted Negative Values: {negative_numbers_list}")
````

# **Task 2: String Character Analysis**

❖ Scenario:

You are building a text-analysis feature.

❖ Task:

Use Gemini to generate a Python function that counts vowels, consonants,
and digits in a string.

❖ Expected Output:

➢ Working function

➢ Sample inputs and outputs

````python
def analyze_string_characters(input_string):
    """Counts vowels, consonants, and digits in a given string."""
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0

    for char in input_string:
        if char.isalpha():  # Check if the character is an alphabet
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
        elif char.isdigit():  # Check if the character is a digit
            digit_count += 1

    return vowel_count, consonant_count, digit_count

# Get user input for string analysis
user_input_string = input("Enter a string for character analysis: ")

v, c, d = analyze_string_characters(user_input_string)

print(f"String: '{user_input_string}'")
print(f"  Vowels: {v}, Consonants: {c}, Digits: {d}")
````

# **Task 3: Palindrome Check – Tool Comparison**

````python
def is_palindrome_gemini(text):
    """
    Checks if a given string is a palindrome.
    Ignores non-alphanumeric characters and is case-insensitive.
    """
    # Clean the text: remove non-alphanumeric characters and convert to lowercase
    cleaned_text = ''.join(char.lower() for char in text if char.isalnum())

    # Check if the cleaned text is equal to its reverse
    return cleaned_text == cleaned_text[::-1]

# Sample inputs for Gemini's function
gemini_test_strings = [
    "madam",
    "A man, a plan, a canal: Panama",
    "hello",
    "Racecar",
    "12321",
    "Was it a car or a cat I saw?",
    "OpenAI"
]

print("--- Gemini's Palindrome Function Tests ---")
for s in gemini_test_strings:
    result = is_palindrome_gemini(s)
    print(f"'{s}': {result}")
print("\n")
````

**Python function (prime check)**

````python
# Prime number check with user input

def is_prime(n: int) -> bool:
    # Handle edge cases
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    # Check divisibility up to sqrt(n)
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Take input from user
num = int(input("Enter a number: "))

# Check and display result
if is_prime(num):
    print(f"✅ {num} is a prime number.")
else:
    print(f"❌ {num} is not a prime number.")
````

**AI Explanation: Line-by-Line Breakdown of `is_prime` function**

Here's a detailed explanation of each significant line in the `is_prime` function:

*   `def is_prime(n: int) -> bool:`
    *   This defines a function named `is_prime`.
    *   `n: int` specifies that the function expects one argument, `n`, which is an integer (this is a type hint).
    *   `-> bool` specifies that the function is expected to return a boolean value (`True` or `False`) (this is also a type hint).

*   `# Handle edge cases`
    *   This is a comment indicating that the following lines deal with special conditions for prime numbers.

*   `if n <= 1:`
    *   This line checks if the input number `n` is less than or equal to 1. By definition, numbers less than or equal to 1 are not prime.

*   `return False`
    *   If `n` is 1 or less, the function immediately returns `False` because it's not a prime number.

*   `if n == 2:`
    *   This line checks if the input number `n` is exactly 2. The number 2 is the smallest and only even prime number.

*   `return True`
    *   If `n` is 2, the function immediately returns `True` because 2 is a prime number.

*   `if n % 2 == 0:`
    *   This line checks if `n` is an even number (i.e., its remainder when divided by 2 is 0). Since 2 is already handled, any other even number cannot be prime.

*   `return False`
    *   If `n` is an even number greater than 2, the function immediately returns `False` because it's not prime.

*   `# Check divisibility up to sqrt(n)`
    *   This comment explains the optimization used in the subsequent loop. We only need to check for divisors up to the square root of `n`.

*   `for i in range(3, int(n**0.5) + 1, 2):`
    *   This starts a `for` loop to check for odd divisors.
    *   `range(3, ..., 2)` means the loop starts checking from 3 and increments by 2 in each step (so it checks only odd numbers).

*   `if n % i == 0:`
    *   Inside the loop, this line checks if `n` is perfectly divisible by the current odd number `i`.

*   `return False`
    *   If `n` is divisible by any `i` in the loop, it means `n` has a factor other than 1 and itself, so it's not prime. The function immediately returns `False`.

*   `return True`
    *   If the loop completes without finding any divisors, it means `n` has no factors other than 1 and itself (after handling edge cases), so it is a prime number. The function returns `True`.
